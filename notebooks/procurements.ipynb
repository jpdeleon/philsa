## 2022

```python
fp = "../data/PhilSA APP Non CSE as of February 9, 2022.pdf"
```

```python
import hashlib

with open(fp, 'rb') as f:
    h1 = hashlib.sha256(f.read()).digest()
    
assert h1.hex()=='583478b65fab0382252585bdba3393005dafc3188c42e636aed0104683ceb3e0'
```

```python
import tabula
import numpy as np

columns = "Code,Project,PMO,Early_Procurement?,Mode,IB-REI,Bids,Award,Signing,Source,Total,MOOE,CO,Remarks".split(',')
df = tabula.read_pdf(fp, 
                     stream=False, 
                     pages="all", 
                     pandas_options={'header': None,
                                     'skiprows': [0,1,2],
                                     'skipfooter': 6,
                                     'engine': 'python'
                                    },
                     output_format='dataframe'
                    )[0]
#cleanup
df = df.drop_duplicates().reset_index(drop=True)

df = df.replace('-', np.nan)
df.columns = columns 
df.tail()
```

```python
set(df.Code)
```

```python
# total1 = df.iloc[155]['Total']
df2 = df.drop([20,21,22,
               107,
#                157,158,159,160
              ],axis=0)
cols = ['Total','MOOE','CO']
df2[cols] = df2[cols].applymap(lambda x: str(x).replace(',','')).applymap(float)

cols2 = ['Project', 'Remarks']
df2[cols2] = df2[cols2].applymap(lambda x: x.replace('\r', ' '))
df2.tail()
```

```python
set(df2.Code)
```

```python
df2.sort_values(by='Total', ascending=False)[['Project','Total','Remarks']].head(20)
```

```python
df2.Total.sum()/1e6
```
